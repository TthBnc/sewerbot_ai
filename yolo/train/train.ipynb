```python
import os
dataset = "11.03"
dataset_version = 'v1i'
root_dir = os.path.dirname(os.path.abspath("__file__"))
config_path = os.path.join(root_dir, "..", "src", dataset, dataset_version, "data.yaml")
print(config_path)
```

```python
from ultralytics import YOLO

# model = YOLO('yolov8n.yaml') # this builds the base model for training
model = YOLO('yolov8n.pt')
```

```python
os.environ['CUDA_LAUNCH_BLOCKING'] = '1'
```

```python
results = model.train(
    data=config_path,
    epochs=250,
    batch=92,
    save_dir='checkpoints/',
    save_period=50,
    name='head_aug',
    seed=42)
```

```python
results = model.val()
```

```python
results = model(asd, conf=0.3)
        
class_names = results[0].names
detections = []
for idx, class_tensor in enumerate(results[0].boxes.cls):
    class_idx = class_tensor.item()
    detections.append({'class': class_names[class_idx], 'prob': results[0].boxes.conf[idx].item()})
detections
```

```python
results = model(test_img, conf=0.3)
class_names = results[0].names
detected_classes = []
for class_tensor in results[0].boxes.cls:
    class_idx = class_tensor.item()
    detected_classes.append(class_names[class_idx])
print(detected_classes)
```
